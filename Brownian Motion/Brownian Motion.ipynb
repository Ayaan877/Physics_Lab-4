```python
# Trimming arrays to the same length
min_len = min(len(l) for l in X_raw)

print(min_len)
    
x = np.array([X[:min_len] for X in X_raw])
y = np.array([Y[:min_len] for Y in Y_raw])
t = np.array([T[:min_len] for T in T_raw])

# Calculating displacement and step length
r = np.zeros(shape=(len(x), min_len))
l = np.zeros(shape=(len(x), min_len-1))

for i in range(len(x)):
    r[i] = np.sqrt(x[i]**2+y[i]**2)
    l[i] = np.diff(r[i])

# Step length histogram
L = l.flatten()

mu = np.mean(L)
std = np.std(L)

lin = np.linspace(-4*std, 4*std, 100)
gauss = np.exp(-(lin-mu)**2/(2*std**2))/(np.sqrt(2*np.pi)*std)

plt.hist(L, bins=40, density=True, label='Step Length Frequency')
plt.plot(lin, gauss, color='red', label='Gaussian Fit')
plt.text(2, 0.3, f'$\mu$ = {np.round(mu, 3)}, $\sigma$ = {np.round(std, 3)}', fontsize=12)
plt.title('Step Length Histogram')
plt.xlabel(r'Step Length [$\mu m$]')
plt.ylabel('Normalised Frequency')
plt.legend()
plt.show()
```

```python
# Diffusion constant (from probability distribution)

D1 = (std**2/2)*1e-12

d = std*np.sqrt(2*12500)

del_D1 = D1*np.sqrt(2*(0.6/d)**2)

# Boltzmann constant (from probability distribution)
T = 273+23.9
R = 0.5e-6
n = 9.1e-4
K = 1.38e-23
k1 = D1*6*np.pi*R*n/T

per_1 = (1-(k1/K))*100
print(per_1)

del_k1 = k1*np.sqrt((del_D1/D1)**2+2*(0.1/T))

print('Diffusion Constant =', D1, f'+- {del_D1}')
print('Boltzmann Constant =', k1, f'+- {del_k1}')
```

```python
# Raw data (for particle ensemble)
x_raw = [x3, x5, x6, x7, x8, x10, x13, x14, x18]
y_raw = [y3, y5, y6, y7, y8, y10, y13, y14, y18]
t_raw = [t3, t5, t6, t7, t8, t10, t13, t14, t18]

# Trimming arrays to the same length
min_len = min(len(l) for l in x_raw)

print(min_len)
    
x = np.array([X[:min_len] for X in x_raw])
y = np.array([Y[:min_len] for Y in y_raw])
t = np.array([T[:min_len] for T in t_raw])

# Calculating displacement and step length
r = np.zeros(shape=(len(x), min_len))
l = np.zeros(shape=(len(x), min_len-1))

for i in range(len(x)):
    r[i] = np.sqrt(x[i]**2+y[i]**2)
    l[i] = np.diff(r[i])
```

```python
# RMS displacement (particle ensemble)
r_rms = np.mean(r**2, axis=0)
r_err = np.std(r**2, axis=0)

cut1 = 1
cut2 = 250

def line(a, x):
    return a*x

par, cov = curve_fit(line, t[0][cut1:cut2], r_rms[cut1:cut2], sigma=r_err[cut1:cut2], absolute_sigma=True)

plt.plot(t[0][cut1:cut2], r[0][cut1:cut2]**2, alpha=0.5, zorder=0, ls='dashed', label=f'$r^2$ Data')
for i in range(1, len(r)):
    plt.plot(t[0][cut1:cut2], r[i][cut1:cut2]**2, alpha=0.5, ls='dashed', zorder=0)

plt.errorbar(t[0][cut1:cut2], r_rms[cut1:cut2], yerr=r_err[cut1:cut2], color='blue', ecolor='grey', alpha=0.8, 
             label='MSD with Standard Deviation')
plt.plot(t[0][cut1:cut2], par[0]*t[0][cut1:cut2], color='red', alpha=0.8, zorder=10, label='Weighted Linear Fit')

# Slope envelope 
s_err = np.sqrt(cov[0][0])
min_s = (par[0]+s_err)*t[0][cut1:cut2]
max_s = (par[0]-s_err)*t[0][cut1:cut2]

plt.plot(t[0][cut1:cut2], min_s, color='orange', ls='--', linewidth=1, alpha=0.8)
plt.plot(t[0][cut1:cut2], max_s, color='orange', ls='--', linewidth=1, alpha=0.8) 
plt.fill_between(t[0][cut1:cut2], min_s, max_s, color='orange', alpha=0.8, zorder=2, label='Fit Envelope')

plt.text(10, 700, f'y = ({np.round(par[0], 2)}$\pm${np.round(s_err, 2)})x', fontsize=12)
plt.title('Particle Ensemble MSD vs Time')
plt.xlabel(r'Time [s]')
plt.ylabel(r'Mean Square Displacement (MSD) [$\mu m^2$]')
plt.legend(loc='upper left')
plt.show()
```

```python
# Diffusion constant (from particle ensemble)

D2 = (par[0]/4)*1e-12

del_D2 = D2*(s_err/par[0])

# Boltzmann constant (from particle ensemble)

k2 = (D2*6*np.pi*n*R)/T

del_k2 = k2*np.sqrt((del_D2/D2)**2+2*(0.1/T))

per_2 = (1-(k2/K))*100
print(per_2)

print(f'Diffusion Constant = {D2} +- {del_D2}')
print(f'Boltzmann Constant = {k2} +- {del_k2}')
```

```python
# RMS displacement (time ensemble)

arr = 11

x_new = []
y_new = []

for i in range(0, len(X_raw[arr]), 10):
    x_new.append(X_raw[arr][i:i+10])
    y_new.append(X_raw[arr][i:i+10])

R_sq = []

for j in range(len(x_new)):
    r_sq = []
    for m in range(len(x_new[0])):
        r0_sq = ((x_new[j][m]-x_new[j][0])**2+(y_new[j][m]-y_new[j][0])**2)
        r_sq.append(r0_sq)
    R_sq.append(r_sq)

Int = np.arange(len(R_sq[0]))
plt.scatter(Int, R_sq[0], s=7, label=f'$r^2$ Data')

for k in range(1, len(R_sq)):
    R_sq[k] = np.abs(R_sq[k]-R_sq[k][0])
    Int = np.arange(len(R_sq[k]))
    plt.scatter(Int, R_sq[k], s=7, alpha=0.5)

R_m = np.mean(R_sq, axis=0)
R_std = np.std(R_sq, axis=0)

def line(a, x):
    return a*x

popt, pcov = curve_fit(line, Int, R_m)

# Fit envelope
S_err = np.sqrt(cov[0][0])
min_S = (popt[0]+S_err)*Int
max_S = (popt[0]-S_err)*Int

plt.plot(Int, min_S, color='orange', ls='-', linewidth=1, alpha=0.8)
plt.plot(Int, max_S, color='orange', ls='-', linewidth=1, alpha=0.8) 
plt.fill_between(Int, min_S, max_S, color='orange', alpha=0.8, zorder=2, label='Fit Envelope')

plt.errorbar(Int, R_m, yerr=R_std, fmt='o', markeredgecolor='black', color='blue', ecolor='k', 
             label=f'MSD with Standard Deviation')
plt.plot(Int, popt[0]*Int, color='red', label=f'Weighted Linear Fit')
plt.text(1, 40, f'y = ({np.round(popt[0], 2)}$\pm${np.round(S_err, 2)})x', fontsize=12)
plt.ylim(-10, 90)
plt.title('Time Ensemble MSD vs Time')
plt.xlabel(r'Time [s]')
plt.ylabel(r'Mean Square Displacement $r_{RMS}^2$ [$\mu m^2$]')
plt.legend()
plt.show()
```
