```python
import numpy as np
import matplotlib.pyplot as plt                     
import scipy as scp
from scipy.optimize import curve_fit
```

```python
V = np.array([140, 160, 180, 200, 220, 240, 260])
D1 = np.array([72.5, 79, 88.5, 91.5, 97.5, 101, 106])*1e-3
D2 = np.array([72.5, 77, 86, 92.5, 97, 101.5, 105])*1e-3
D3 = np.array([72.5, 81, 85, 93, 95.5, 99.5, 102])*1e-3

D_err = np.std([D1, D2, D3], axis=0)

D = (D1+D2)/2

Dsq_err = ((D+D_err)**2-(D-D_err)**2)/2

a, b = np.polyfit(V, D**2, 1, cov=True)
print(np.sqrt(b[0][0]))
#plt.scatter(V, D**2, label='Data')
plt.errorbar(V, D**2, yerr=Dsq_err, ecolor='black', fmt='o', label='Data')
plt.plot(V, a[0]*V+a[1], color='red', label='Fit')
print(a[0])
plt.legend()
plt.xlabel(r'Accelerating Voltage ($V$) [V]')
plt.ylabel(r'Diameter$^2$ ($D^2$) [m$^2$]')
plt.text(155, 0.009, 'slope = '+str(np.round(a[0], 8)), fontsize=12)
plt.show()
```

```python
EM = 1.75e11

em = 9.2e6/a[0]

Error = (EM-em)/EM

print(em*1e-11)
print(Error)
```
