```python
f = 65e-3
w = 546.1e-9
R0 = r0*14e-6

t = []

for i in range(len(R0)-1):
    n0 = (2*(f**2)*(i+1))/((R0[i+1]**2)-(R0[0]**2))
    T = (n0*w)/(2*1.5)
    t.append(T)

t = np.array(t)

T = t.flatten()

t_mean = np.mean(T)
t_std = np.std(T)

x = np.linspace(t_mean-3*t_std, t_mean+3*t_std, 100)
gauss = np.exp(-(x-t_mean)**2/(2*t_std**2))/(np.sqrt(2*np.pi)*t_std)

plt.plot(x, gauss, color='red')
plt.xlabel('Etalon Spacing [m]')
plt.ylabel('Normalised Frequency')
plt.text(2.75e-3, 8e3, r'mean = '+str(np.round(t_mean, 5)*1e3)+' mm', fontsize=12)
plt.text(2.75e-3, 7e3, r'std = '+str(np.round(t_std, 5)*1e3)+' mm', fontsize=12)
plt.hist(T, bins=12, density=True)
plt.show()
```

```python
I_B = np.array([0.7, 0.8, 0.9, 1.0, 1.1, 1.2, 1.3, 1.4, 1.5, 1.6, 1.7, 1.8, 1.9])
B = 0.306*I_B-0.079

r_a = np.array([r4_a, r5_a, r6_a])*14e-6
r_b = np.array([r4_b, r5_b, r6_b])*14e-6

d = []
D = []

for i in range(len(r_a)):
    delta = np.square(r_b[i]) - np.square(r_a[i])
    Delta_a = np.diff(r_a[i]**2, axis=0)
    Delta_b = np.diff(r_b[i]**2, axis=0)
    
    Delta = (Delta_a + Delta_b)/0.2
    
    Delta = np.mean(Delta, axis=0)
    delta = np.mean(delta, axis=0)
    
    d.append(delta)
    D.append(Delta)

d = np.array(d)
D = np.array(D)

d_ratio = d/D

d_ratio_mean = np.mean(d_ratio, axis=0)
d_ratio_std = np.std(d_ratio, axis=0)

def f(a, x):
    return a*x

a, b = curve_fit(f, B, d_ratio_mean)

mu_err = np.sqrt(np.diag(b))

print('slope error =', mu_err[0])

plt.plot(B, f(B, *a), color='red')
plt.errorbar(B, d_ratio_mean, yerr=d_ratio_std, fmt='o', ecolor='black')
plt.text(0.25, 0.055, 'y = '+str(np.round(a[0], 3))+'x', fontsize = 12)
plt.ylabel(r'Delta Ratio [$\delta_{ab}/\Delta$]')
plt.xlabel('Magnetic Field [T]')
plt.show()
```

```python
h = 6.626e-34
c = 3*10e8

mu = (h*c*a[0])/(t_mean)
mu_up = (h*c*(a[0]+mu_err[0]))/(t_mean)
mu_low = (h*c*(a[0]-mu_err[0]))/(t_mean)

print(mu-mu_low)
print(mu_up-mu)

print(mu, mu_up, mu_low)
```
