```python
import numpy as np
from numpy import polyval
import matplotlib.pyplot as plt
import scipy as scp  
from scipy.optimize import curve_fit
```

```python
res, v_sq, v_sq_std, g2 = np.loadtxt('../Noise Fundamentals/Measuring Johnson Noise.csv', skiprows=4, usecols=(0, 1, 2, 3), delimiter=',', unpack=True)
R = res[0:]
V_sq = v_sq[0:]
V_sq_std = v_sq_std[0:]
G2 = g2[0:]
```

```python
# Measuring Johnson Noise
# R_f=1kOhm, C_f=min, HP_filter=0.1kHz, LP_filter=100kHz, Time_avg=1s, Meter=0-2V
G1 = 600

V_jn = V_sq*10/(G1*G2)**2
V_std = V_sq_std*10/(G1*G2)**2

def line(a, b, x):
    return a*x+b

par1, cov1 = curve_fit(line, R, V_jn, sigma=V_std, absolute_sigma=True)

Vn_err = np.sqrt(cov1[0][0])

V_n = par1[0]

print(f'V_n = {V_n}+-{Vn_err} V')

r = np.linspace(1, R[0], 10000)

plt.errorbar(R, V_jn, yerr=V_std*100, ms=4, ecolor='k', color='teal', fmt='o', 
             label=r'$\langle V^2_j+V^2_N\rangle$')
plt.plot(r, line(r, *par1), color='mediumvioletred', label='Weighted Linear Fit')
plt.text(10000, 1e-10, f'y = ({np.round(par1[1], 17)})x+({np.round(par1[0], 14)})')
plt.xlabel(r'Input Resistance [$\Omega$]')
plt.ylabel(r'Mean Square Voltage $\langle V^2_j + V^2_N\rangle$ [V$^2$]')
plt.legend()
plt.show()
```

```python
V_j = V_jn - V_n

def line2(a, x):
    return a*x

par2, cov2 = curve_fit(line2, R, V_j, sigma=V_std, absolute_sigma=True)

band = 110961
T = 300
K_real = 1.38e-23

s1 = 4*band*T*K_real

plt.errorbar(np.log(R), V_j, yerr=V_std*100, ms=4, ecolor='k', color='teal', fmt='o', label=r'$\langle V^2_j\rangle$')
plt.plot(np.log(r), r*par2[0], color='r', label='Data Trend')
plt.plot(np.log(r), r*s1, color='g', ls='--', label='Expected Trend')
plt.ylim(-0.01e-9, 0.35e-9)
#plt.text(30000, 0.9e-10, f'y = ({np.round(par2[0], 17)})x')
plt.xlabel(r'Log(Input Resistance) [$\ln(\Omega)$]')
plt.ylabel(r'Mean Square Voltage $\langle V^2_j\rangle$ [V$^2$]')
plt.legend()
plt.show()
```

```python
slope1 = par2[0]

slope1_err = np.sqrt(cov2[0][0])

band = 110961
T = 300

K1 = slope1/(4*band*T)

K1_err = slope1_err/(4*band*T)

K_real = 1.38e-23

K1_per_err = np.abs(1-K1/K_real)*100

print(f'Boltzmann Constant = {K1}+-{K1_err} J/K')
print(f'Percentage Error = {K1_per_err} %')
```

```python
del_f, V_jn1, V_std1 = np.loadtxt('../Noise Fundamentals/f2-f1.csv', skiprows=1, delimiter=',', unpack=True)
B, V_jn2, V_std2 = np.loadtxt('../Noise Fundamentals/Bandwidth.csv', skiprows=1, delimiter=',', unpack=True)
```

```python
# Bandwidth
# R_in=10kOhm, R_f=1kOhm, C_f=min, Time_avg=1s, Meter=0-2V

w1 = 1/V_std1
w2 = 1/V_std2

p1, c1 = np.polyfit(del_f, V_jn1, 1, cov=True)
p2, c2 = np.polyfit(B, V_jn2, 1, w=w2, cov=True)

V_n1 = p1[1]
V_n2 = p2[1]

V_j1 = V_jn1 - V_n1
V_j2 = V_jn2 - V_n2

b1 = np.linspace(1, 1e5, 1000)
b2 = np.linspace(0.4, 1.11061e+05, 1000)

plt.errorbar(del_f, V_j1, yerr=V_std1*40, ms=4, fmt='o', ecolor='k', label=f'$|f_2-f_1|$')
plt.errorbar(B, V_j2, yerr=V_std2*40, ms=4, fmt='o', ecolor='k', label='Effective Bandwidth')
plt.plot(b1, polyval(p1, b1), label=f'$|f_2-f_1|$ Weighted Linear Fit')
plt.plot(b2, polyval(p2, b2), label='Effective Bandwidth Weighted Linear Fit')
plt.text(25000, 1.5e-11, f'y = ({np.round(p1[0], 18)})x')
plt.text(60000, 1e-11, f'y = ({np.round(p2[0], 18)})x')
plt.xlabel(r'Bandwidth ($\Delta f$) [Hz]')
plt.ylabel(r'Voltage $\langle V^2_j\rangle$ [V$^2$]')
plt.legend()
plt.show()
```

```python
slope2 = p1[0]
slope3 = p2[0]
slope2_err = np.sqrt(c1[0][0])
slope3_err = np.sqrt(c2[0][0])
R_in = 10000
T = 300

K2 = slope2/(4*R_in*T)
K3 = slope3/(4*R_in*T)

K2_err = slope2_err/(4*R_in*T)
K3_err = slope3_err/(4*R_in*T)

K2_per_err = np.abs(1-K2/K_real)*100
K3_per_err = np.abs(1-K3/K_real)*100

print(f'Boltzmann Constant from |f2-f1| = {K2}+-{K2_err} J/K')
print(f'Percentage Error from |f2-f1| = {K2_per_err} %')
print(f'Boltzmann Constant from Bandwidth = {K3}+-{K3_err} J/K')
print(f'Percentage Error from Bandwidth = {K3_per_err} %')
```

```python
temp, V_t = np.loadtxt('../Noise Fundamentals/Temperature Calibration.csv', skiprows=2, delimiter=',', unpack=True)
V_t = V_t/1000
```

```python
#Temperature Calibration
p3, c3 = np.polyfit(V_t, temp, 1, cov=True)

plt.scatter(V_t, temp, label='Temperature')
plt.plot(V_t, polyval(p3, V_t), color='r', label='Linear Fit')
plt.text(0.3, 200, f'y = {np.round(p3[0], 2)}x + {np.round(p3[1], 2)}')
plt.xlabel('Diode Voltage [V]')
plt.ylabel('Probe Temperature [K]')
plt.legend()
plt.show()
```

```python
V_T, V_jn3, V_std3 = np.loadtxt('../Noise Fundamentals/Temperature.csv', skiprows=3, usecols=(0, 4, 5), delimiter=',', unpack=True)
```

```python
#Temperature Dependence
Temp = p3[0]*V_T+p3[1]

w3 = 1/V_std3

p4, c4 = np.polyfit(Temp, V_jn3, 1, w=w3, cov=True)

V_n3 = p4[1]

V_j3 = V_jn3 - V_n3

plt.errorbar(Temp, V_j3, yerr=V_std3*10, ms=4, fmt='o', color='teal', ecolor='k', label=r'$\langle V^2_j\rangle$')
plt.plot(Temp, p4[0]*Temp, color='mediumvioletred', label='Weighted Linear Fit', zorder=10)
plt.text(200, 0.9e-12, f'y = ({np.round(p4[0], 17)})x')
plt.xlabel(r'Temperature [K]')
plt.ylabel(r'Mean Square Voltage $\langle V^2_j\rangle$ [V$^2$]')
plt.legend()
plt.show()
```

```python
s = p4[0]
s_err = np.sqrt(c4[0][0])
R_in = 10000
bandwidth = 9997

K4 = s/(4*R_in*bandwidth)

K4_err = s_err/(4*R_in*bandwidth)

K4_per_err = np.abs(1-K4/K_real)*100

print(f'Boltzmann Constant from Temperature = {K4}+-{K4_err} J/K')
print(f'Percentage Error from Temperature = {K4_per_err} %')
```
